## Limpando e separando dados

```python
import pandas as pd

def clean_data(df):
    # Alterar tipo de coluna para datetime64[ns] para coluna: 'release_date'
    df = df.astype({'release_date': 'datetime64[ns]'})
    # Alterar tipo de coluna para category para coluna: 'owners'
    df = df.astype({'owners': 'category'})
    # Dividir texto usando a cadeia de caracteres ';' nas colunas: 'steamspy_tags', 'categories', 'platforms'
    loc_0 = df.columns.get_loc('steamspy_tags')
    df_split = df['steamspy_tags'].str.split(pat=';', expand=True).add_prefix('steamspy_tags_')
    df = pd.concat([df.iloc[:, :loc_0], df_split, df.iloc[:, loc_0:]], axis=1)
    loc_1 = df.columns.get_loc('categories')
    df_split = df['categories'].str.split(pat=';', expand=True).add_prefix('categories_')
    df = pd.concat([df.iloc[:, :loc_1], df_split, df.iloc[:, loc_1:]], axis=1)
    loc_2 = df.columns.get_loc('platforms')
    df_split = df['platforms'].str.split(pat=';', expand=True).add_prefix('platforms_')
    df = pd.concat([df.iloc[:, :loc_2], df_split, df.iloc[:, loc_2:]], axis=1)
    # df = df.drop(columns=['steamspy_tags', 'categories', 'platforms'])
    return df

# df = pd.read_csv('steamKagleDataPkg\steam.csv')

df_clean = clean_data(df.copy())
df_clean.head()
```
