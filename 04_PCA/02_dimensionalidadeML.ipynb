### 4. Reduzir dimensionalidade através da análise de componentes principais (PCA)

#### 4.1. O que é uma projeção linear aplicada ao espaço de dados
Uma **projeção linear** é uma transformação que mapeia um conjunto de pontos de um espaço dimensional para outro, preservando a linearidade das relações entre os pontos. Em PCA, a projeção linear é utilizada para transformar os dados originais em um novo conjunto de eixos (componentes principais) de forma que a variância dos dados ao longo desses eixos seja maximizada.

#### 4.2. Etapas do algoritmo de PCA para extração dos componentes principais
O algoritmo de PCA segue as seguintes etapas principais:
1. **Centralização dos Dados**: Subtrair a média de cada variável para que cada variável tenha uma média zero.
2. **Cálculo da Matriz de Covariância**: Calcular a matriz de covariância para entender como as variáveis se relacionam entre si.
3. **Cálculo dos Autovalores e Autovetores**: Derivar os autovalores e autovetores da matriz de covariância. Os autovetores são os novos eixos (componentes principais) e os autovalores indicam a importância de cada componente.
4. **Ordenação dos Autovetores**: Ordenar os autovetores em ordem decrescente de seus autovalores.
5. **Projeção dos Dados**: Projetar os dados originais sobre os novos eixos principais.



#### 4.3. Utilizar PCA em uma base de dados e selecionar as componentes mais relevantes

```python
from sklearn.decomposition import PCA
import numpy as np

# Gerar dados de exemplo
# np.random.seed(42)
data = np.random.rand(100, 10)

# Aplicar PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data)

print("Componentes Principais:\n", principal_components)
print("Variância Explicada:\n", pca.explained_variance_ratio_)
```

#### 4.4. Relacionar o processo de whitening e PCA
**Whitening** é uma transformação adicional aplicada após a PCA que escala os componentes principais para que tenham variância unitária e sejam decorrelacionados. Isto é útil para alguns algoritmos que assumem que os dados de entrada têm variância unitária.

#### 4.5. Escolher o número de componentes de uma PCA baseado na curva de carga
A **curva de carga**, ou gráfico de scree, é uma ferramenta visual utilizada para determinar o número adequado de componentes principais. A curva plota a variância explicada por cada componente principal em ordem decrescente. O "joelho" da curva, onde a taxa de variação começa a diminuir, é frequentemente utilizado como um ponto de corte para selecionar o número de componentes.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

# Gerar dados de exemplo
np.random.seed(42)
data = np.random.rand(100, 150)

# Aplicar PCA
pca = PCA()
pca.fit(data)

# Plotar a curva de carga
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Explicada Acumulada')
plt.title('Curva de Carga')
plt.grid(True)
plt.show()
```

#### 4.6. Limitações da PCA
- **Assume Linearidade**: PCA só captura relações lineares entre as variáveis, podendo não ser eficaz para dados com relações não lineares.
- **Sensibilidade ao Escalonamento**: PCA é sensível à escala das variáveis, por isso é importante normalizar os dados antes de aplicar a PCA.
- **Interpretação das Componentes**: As componentes principais podem ser difíceis de interpretar em termos de variáveis originais.
- **Perda de Informação**: Ao reduzir a dimensionalidade, alguma informação pode ser perdida, especialmente se muitos componentes forem descartados.

#### 4.7. Aplicar PCA em diversas bases de diferentes contextos
Podemos aplicar PCA a diferentes tipos de dados, como imagens, dados financeiros, ou dados biológicos, para reduzir a dimensionalidade e facilitar a visualização ou melhorar o desempenho de algoritmos de machine learning.

```python
from sklearn.datasets import load_iris
import pandas as pd

# Carregar dados de exemplo (Iris dataset)
iris = load_iris()
data = iris.data

# Aplicar PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data)

# Criar DataFrame com componentes principais
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
print(principal_df.head())
```

### 6. Reduzir dimensionalidade através da Análise de Componentes Independentes (ICA)

#### 6.1. O que é a Análise de Componentes Independentes (ICA)
A **Análise de Componentes Independentes (ICA)** é uma técnica de processamento de sinais e estatística que busca encontrar componentes independentes em um conjunto de sinais misturados. Ao contrário da PCA, que busca maximizar a variância e encontrar componentes ortogonais, a ICA busca componentes que são estatisticamente independentes.

#### 6.2. Diferenças entre ICA e PCA
- **Objetivo**: ICA busca independência estatística, enquanto PCA busca maximizar a variância e ortogonalidade.
- **Aplicação**: ICA é usada frequentemente em processamento de sinais, como na separação de fontes em misturas auditivas (problema do coquetel).
- **Independência**: ICA resulta em componentes que são independentes, enquanto PCA resulta em componentes não correlacionados.

#### 6.3. Aplicar ICA em uma base de dados
Vamos aplicar ICA em um conjunto de dados e visualizar os componentes independentes.

```python
import numpy as np
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt

# Gerar dados de exemplo (mistura de sinais sinusoidais)
np.random.seed(0)
time = np.linspace(0, 10, 2000)
s1 = np.sin(2 * time)  # Sinal 1: seno
s2 = np.sign(np.sin(3 * time))  # Sinal 2: onda quadrada
S = np.c_[s1, s2]
S += 0.2 * np.random.normal(size=S.shape)  # Adicionar ruído
S /= S.std(axis=0)  # Padronizar

# Misturar os sinais
A = np.array([[1, 1], [0.5, 2]])  # Matriz de mistura
X = np.dot(S, A.T)  # Misturar os sinais para obter observações

# Aplicar ICA
ica = FastICA(n_components=2)
S_ = ica.fit_transform(X)  # Reconstruir os sinais
A_ = ica.mixing_  # Estimar a matriz de mistura

# Plotar os sinais originais, misturados e reconstruídos
plt.figure(figsize=(9, 6))
plt.subplot(3, 1, 1)
plt.title('Sinais Originais')
plt.plot(S)
plt.subplot(3, 1, 2)
plt.title('Sinais Misturados')
plt.plot(X)
plt.subplot(3, 1, 3)
plt.title('Sinais Reconstituídos')
plt.plot(S_)
plt.tight_layout()
plt.show()
```

#### 6.4. Explicar as limitações da ICA
- **Sensibilidade ao Ruído**: ICA pode ser sensível ao ruído nos dados, o que pode afetar a independência dos componentes extraídos.
- **Número de Componentes**: O número de componentes independentes deve ser especificado previamente, o que pode ser um desafio em aplicações práticas.
- **Assumir Independência**: ICA assume que os sinais de origem são estatisticamente independentes, o que pode não ser verdade em todas as situações.
