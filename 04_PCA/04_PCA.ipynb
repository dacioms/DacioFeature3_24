## O que é uma projeção linear aplicada ao espaço de dados

A projeção linear é uma operação matemática que mapeia pontos de um espaço de alta dimensão para um espaço de dimensões mais baixas, preservando ao máximo as características estruturais dos dados. Em termos práticos, no contexto de redução de dimensionalidade, ela visa projetar os dados em um novo sistema de coordenadas, onde as direções das projeções são escolhidas de modo a capturar a maior quantidade possível de variação dos dados originais.

No caso do PCA, essa projeção linear é realizada através da escolha de novos eixos (ou componentes) que maximizam a variação dos dados. Cada componente principal é uma combinação linear dos atributos originais, e as projeções são feitas ao longo desses novos eixos, que representam direções de maior variação no conjunto de dados.



### As etapas do algoritmo de PCA para extração dos componentes principais

1. **Centralização dos dados**: Subtrai-se a média de cada variável dos dados, de modo que o conjunto fique centrado em torno da origem. Isso garante que o PCA avalie apenas as variações e não a posição absoluta dos dados.
   
2. **Cálculo da matriz de covariância**: Com os dados centralizados, é necessário calcular a matriz de covariância para entender como os diferentes atributos dos dados estão correlacionados entre si. A matriz de covariância descreve a variabilidade conjunta entre os diferentes atributos.

3. **Cálculo dos autovalores e autovetores**: A matriz de covariância é decomposta em autovalores e autovetores. Os autovetores representam as direções das novas dimensões (componentes principais) e os autovalores indicam a quantidade de variância explicada por cada uma dessas direções.

4. **Ordenação dos autovetores**: Os autovetores são ordenados de acordo com seus autovalores, de modo que o autovetor associado ao maior autovalor se torne o primeiro componente principal, capturando a maior variação possível dos dados. O segundo autovetor, associado ao segundo maior autovalor, se torna o segundo componente principal, e assim por diante.

5. **Projeção dos dados**: Finalmente, os dados originais são projetados no espaço dos componentes principais, utilizando os autovetores como novos eixos de coordenadas.



### Utilizar PCA em uma base de dados e selecionar as componentes mais relevantes

Ao aplicar PCA a uma base de dados, o objetivo é reduzir a dimensionalidade, preservando a maior parte possível da variabilidade dos dados. Após a execução do algoritmo, cada componente principal estará associado a uma quantidade de variação explicada. Para selecionar as componentes mais relevantes, pode-se aplicar os seguintes critérios:

1. **Variância explicada acumulada**: A soma das variâncias explicadas por cada componente pode ser usada para decidir o número de componentes a manter. Por exemplo, um critério comum é manter componentes que expliquem, cumulativamente, pelo menos 95% da variância.

2. **Gráfico de curva de carga (scree plot)**: Esse gráfico exibe a variância explicada por cada componente. O ponto onde a variância adicional começa a decrescer drasticamente indica o número ótimo de componentes a manter, baseado no conceito de "cotovelo" do gráfico.



### Relacionar o processo de whitening e PCA

Whitening é um processo aplicado aos dados após a aplicação do PCA para eliminar qualquer correlação residual entre as componentes principais. O objetivo do whitening é transformar os dados de tal maneira que todas as suas variáveis tenham variância unitária e sejam incorrelacionadas.

Em termos práticos, após projetar os dados nas componentes principais, os componentes ainda podem ter diferentes escalas. O whitening normaliza esses componentes, de modo que eles tenham uma variância igual a 1. Esse processo é útil em alguns algoritmos de aprendizado de máquina que requerem dados com essas propriedades, embora, em algumas aplicações, o whitening possa aumentar o ruído.



### Escolher o número de componentes de uma PCA baseado na curva de carga

A curva de carga (scree plot) é uma ferramenta visual que mostra a variância explicada por cada componente principal. A escolha do número de componentes pode ser baseada no ponto de inflexão da curva, também conhecido como "cotovelo". A ideia é identificar o número mínimo de componentes que retêm a maior parte da variância dos dados, antes que o ganho marginal de variância explicada comece a diminuir de forma significativa.



## Limitações da PCA

1. **Suposição de linearidade**: PCA pressupõe que as relações entre variáveis são lineares. Se os dados apresentarem relações não lineares significativas, o PCA não será eficiente para capturar essas variações.

2. **Sensibilidade à escala das variáveis**: Como o PCA depende da variância dos dados, é sensível à escala das variáveis. É importante normalizar os dados antes de aplicar o PCA, caso contrário, variáveis com grandes magnitudes podem dominar a análise.

3. **Perda de interpretabilidade**: As componentes principais são combinações lineares dos atributos originais, o que pode dificultar a interpretação dos novos eixos em termos de características reais dos dados.

4. **Eficiência com variância alta**: O PCA funciona melhor quando as variáveis que mais contribuem para a variância dos dados são de interesse. Em situações onde as variáveis mais importantes apresentam baixa variância, o PCA pode não ser uma boa escolha.

Aplicar PCA utilizando a biblioteca Scikit-learn.

Aqui, consideramos um exemplo simples de aplicação em um conjunto de dados fictício.

### Passos:

1. **Instalar e importar as bibliotecas necessárias**:


2. **Importação de bibliotecas e preparação dos dados**:
   No exemplo, usamos a biblioteca `Pandas` para carregar os dados e a biblioteca `numpy` para manipulação de arrays, além da `sklearn.decomposition.PCA` para a aplicação da PCA.



```python

import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```



3. **Carregar os dados e padronizar (normalizar)**:
   Antes de aplicar o PCA, os dados precisam ser normalizados para garantir que as variáveis estejam na mesma escala.



```python

# Exemplo fictício de dados (7 disciplinas)
dados = {'Matemática': [85, 78, 92, 88, 76],
         'Física': [82, 76, 91, 84, 72],
         'Química': [87, 80, 95, 90, 74],
         'Biologia': [78, 74, 88, 83, 70],
         'História': [74, 70, 85, 80, 68],
         'Geografia': [73, 69, 84, 79, 67],
         'Inglês': [90, 88, 94, 92, 86]}

df = pd.DataFrame(dados)

# Normalizando os dados
scaler = StandardScaler()
dados_normalizados = scaler.fit_transform(df)
dados_normalizados
```



4. **Aplicar o PCA**:
   Agora que os dados estão normalizados, podemos aplicar o PCA e determinar quantas componentes queremos manter.



```python
df
```

```python
pd.DataFrame(dados_normalizados)
```

```python

# Definindo o número de componentes que queremos manter (2 neste exemplo)
pca = PCA(n_components=4)

# Ajustar e transformar os dados
componentes_principais = pca.fit_transform(dados_normalizados)

# Exibir os componentes principais
print(componentes_principais)
```


5. **Análise da variância explicada**:
   Podemos verificar quanta variância foi explicada por cada componente principal.


```python
# Variância explicada por cada componente
print(pca.explained_variance_ratio_)

# Variância acumulada
print(np.cumsum(pca.explained_variance_ratio_))
```


6. **Plotando a curva de carga (scree plot)**:
   Um gráfico comum para decidir o número de componentes principais é o scree plot.

   


```python
import matplotlib.pyplot as plt

# Variância explicada
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Explicada Acumulada')
plt.title('Curva de Carga (Scree Plot)')
plt.grid(True)
plt.show()
  
```

## Exemplo de aplicação de PCA na análise de dados de imagens

O PCA também é amplamente utilizado em análise de imagens para reduzir a dimensionalidade, especialmente em compressão de imagens. Vamos considerar um exemplo em que aplicamos PCA a uma imagem em escala de cinza.

### Exemplo com PCA aplicado em imagens:

1. **Carregar e pré-processar a imagem**:
   Vamos começar carregando uma imagem em escala de cinza e convertendo-a em uma matriz.

   


```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from matplotlib import image as mpimg
import matplotlib.pyplot as plt

# Carregar a imagem
img = mpimg.imread('logoInfnet.jpg')

# Converter para escala de cinza (se necessário)
img_gray = np.mean(img, axis=2)

# Mostrar a imagem original
plt.imshow(img_gray, cmap='gray')
plt.title('Imagem Original')
plt.show()
len(img_gray)
```


2. **Normalizar a imagem**:
   Assim como em dados tabulares, é uma boa prática normalizar a imagem antes de aplicar o PCA.



```python

# Normalizando a imagem
img_normalizada = StandardScaler().fit_transform(img_gray)
plt.imshow(img_normalizada, cmap='gray')
plt.title('Imagem Normalizada')
plt.show()
```



3. **Aplicar o PCA**:
   Agora, aplicamos o PCA e determinamos o número de componentes principais que desejamos manter.



```python

# Aplicando PCA e mantendo 50 componentes principais
pca = PCA(50)
img_pca = pca.fit_transform(img_normalizada)

# Projetando a imagem de volta para o espaço original com os componentes principais
img_reconstruida = pca.inverse_transform(img_pca)

# Mostrar a imagem reconstruída
plt.imshow(img_reconstruida, cmap='gray')
plt.title('Imagem Reconstruída com PCA (50 componentes)')
plt.show()
img_reconstruida
```

```python
img_whitened = img_reconstruida / np.sqrt(pca.explained_variance_).reshape(1,1)
plt.imshow(img_whitened, cmap='gray')
plt.title('Imagem Reconstruída esbranqui')
plt.show()
```


4. **Análise da compressão**:
   Ao utilizar um número reduzido de componentes principais (por exemplo, 50), conseguimos comprimir a imagem sem perder muita informação visual. Este é o princípio usado em técnicas de compressão de imagem, como JPEG.

   O processo de reconstrução da imagem após a aplicação do PCA preserva as características principais da imagem, mas com uma representação em menor dimensão. Isso pode ser especialmente útil quando lidamos com grandes bases de imagens, como em reconhecimento facial ou processamento de vídeos.


## Detalhamento do processo de **Whitening** e sua relação com o **PCA**

O **Whitening** é uma técnica complementar frequentemente usada após a aplicação da **Análise de Componentes Principais (PCA)** para garantir que os dados projetados sejam **descorrelacionados** e que todas as suas variáveis tenham **variância unitária**. Vamos detalhar o processo e sua relação com o PCA:

### Entendendo o Whitening

Após a aplicação da PCA, os dados projetados nos componentes principais ainda podem ter diferentes escalas de variância. Isso ocorre porque os componentes principais são baseados na variação dos dados ao longo de diferentes direções, e essas variações são mantidas na forma original.

O **Whitening** visa transformar os dados de tal maneira que:

1. Todas as variáveis resultantes sejam **descorrelacionadas**, ou seja, não haja relação linear entre elas;
2. Todas as variáveis tenham a **mesma variância**, usualmente normalizada para 1.

Em termos práticos, isso significa que as novas variáveis estão na forma **"esbranquiçada"**, ou seja, são como uma "cópia" dos dados originais, mas normalizados para terem variância unitária e zero correlação entre si.

### Por que usar Whitening?

- **Descorrelação**: Após aplicar o PCA, as direções dos componentes principais podem estar descorrelacionadas, mas os dados projetados ao longo dessas direções podem ainda conter alguma correlação. O Whitening remove essa correlação residual, tornando os dados totalmente independentes uns dos outros.

- **Normalização da variância**: Cada componente principal retém uma quantidade diferente de variância dos dados. Com o Whitening, essa variância é ajustada para ser a mesma (tipicamente 1) para todos os componentes, o que pode ser útil em algoritmos de aprendizado de máquina que são sensíveis a variâncias diferentes entre variáveis.

### Como o Whitening é aplicado após o PCA

1. **Projeção no espaço dos componentes principais**: 
   Primeiramente, o PCA é aplicado nos dados, projetando-os em um novo espaço com base nos componentes principais. Cada novo componente pode ter uma variância diferente, que está diretamente relacionada ao autovalor correspondente daquele componente.

2. **Divisão pela raiz dos autovalores**: 
   O Whitening envolve ajustar essa variância. Isso é feito dividindo cada componente principal pela raiz quadrada do autovalor correspondente. Ao fazer isso, cada componente terá variância 1, e todas as variáveis serão ajustadas para terem a mesma escala.

3. **Matriz de transformação**: 
   A operação de Whitening pode ser descrita como a multiplicação dos dados projetados nos componentes principais por uma matriz de transformação, que é composta pelos autovetores do PCA, escalados pelos inversos das raízes quadradas dos autovalores.

### Relação com o PCA

- O PCA transforma os dados em um novo sistema de coordenadas onde a variância ao longo de cada eixo (componente principal) é maximizada.
- No entanto, essas variâncias podem ser muito diferentes entre os componentes principais. O Whitening ajusta essas variâncias para garantir que cada componente principal tenha a mesma contribuição, com uma variância de 1.
- Portanto, o Whitening pode ser visto como uma forma de padronizar o resultado do PCA, mantendo os dados transformados sem correlação, mas ajustados para uma escala comum.

### Exemplo Prático de Whitening em Python (após o PCA)

Aqui está um exemplo prático de como aplicar Whitening após realizar o PCA usando a biblioteca Scikit-learn em Python.

#### 1. Aplicar o PCA

Primeiro, vamos realizar o PCA em um conjunto de dados e projetá-los nos componentes principais.



```python

import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Exemplo fictício de dados
dados = {'Matemática': [85, 78, 92, 88, 76],
         'Física': [82, 76, 91, 84, 72],
         'Química': [87, 80, 95, 90, 74],
         'Biologia': [78, 74, 88, 83, 70],
         'História': [74, 70, 85, 80, 68],
         'Geografia': [73, 69, 84, 79, 67],
         'Inglês': [90, 88, 94, 92, 86]}

df = pd.DataFrame(dados)

# Normalizando os dados
scaler = StandardScaler()
dados_normalizados = scaler.fit_transform(df)

# Aplicando PCA
pca = PCA(n_components=4)
dados_pca = pca.fit_transform(dados_normalizados)
```



#### 2. Aplicar o Whitening

Após aplicar o PCA, fazemos o Whitening dividindo cada componente principal pela raiz do autovalor correspondente.



```python

# Whitening dos dados após o PCA
dados_whitened = dados_pca / np.sqrt(pca.explained_variance_)

# Exibindo os dados whitened
print(dados_whitened)
```



#### 3. Comparar os resultados

Aqui estão os dados projetados pelos componentes principais (PCA) e os dados após o Whitening. No Whitening, todas as variáveis terão variância unitária, enquanto no PCA original, as variâncias são proporcionais aos autovalores dos componentes principais.

### Cuidados com o Whitening

Embora o Whitening possa ser útil, ele também pode ter desvantagens:

1. **Amplificação de ruído**: Ao normalizar as variâncias, o Whitening pode aumentar o impacto de componentes com pouca variação nos dados. Isso pode resultar na amplificação do ruído, especialmente em componentes que inicialmente têm variação pequena e podem não ser relevantes.

2. **Redução da interpretabilidade**: O PCA já pode resultar em componentes principais difíceis de interpretar, e o Whitening pode tornar essa interpretação ainda mais complexa, uma vez que todas as variáveis estão agora em uma escala artificialmente ajustada.

3. **Necessidade específica**: Em muitas aplicações de aprendizado de máquina, o Whitening não é necessário e pode até ser contraproducente, especialmente em dados que já foram suficientemente processados com técnicas como normalização.

### Conclusão sobre Whitening e PCA

- O **PCA** reduz a dimensionalidade dos dados, preservando as direções de máxima variância. 
- O **Whitening** complementa o PCA ao normalizar os dados projetados, garantindo que todas as variáveis tenham variância unitária e estejam completamente descorrelacionadas.
- É uma ferramenta útil em algoritmos que requerem dados sem correlação residual e variância padronizada, mas deve ser aplicada com cautela para evitar amplificação de ruído ou perda de interpretabilidade.



## Exemplo avançado de aplicação de **PCA** e **Whitening** em dados financeiros

Vamos agora explorar um exemplo avançado onde aplicamos a **Análise de Componentes Principais (PCA)** e o **Whitening** a um conjunto de dados financeiros. Este tipo de aplicação é útil em situações onde temos muitas variáveis correlacionadas, como diferentes ativos de mercado, e desejamos reduzir a dimensionalidade sem perder a essência das variações nos dados.

### Contexto

Neste exemplo, vamos analisar os retornos diários de diferentes ações de uma bolsa de valores. O objetivo é aplicar o PCA para identificar os fatores principais que influenciam a variação do mercado e, em seguida, aplicar o Whitening para garantir que os componentes principais estejam descorrelacionados e normalizados.

### Passos para o exemplo:

#### 1. Carregar e normalizar os dados financeiros

Primeiro, carregamos um conjunto de dados fictício que contém o retorno diário de diferentes ações. Esse tipo de dado pode ser obtido de fontes reais, como a API do Yahoo Finance, mas aqui usaremos um exemplo simplificado.



```python

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Dados financeiros fictícios (retornos diários de 5 ações)
dados_financeiros = {'Ação_A': [0.01, 0.02, -0.01, 0.03, -0.02],
                     'Ação_B': [-0.02, 0.03, 0.01, -0.01, 0.02],
                     'Ação_C': [0.03, -0.01, 0.02, -0.03, 0.01],
                     'Ação_D': [0.02, 0.01, -0.02, 0.03, -0.01],
                     'Ação_E': [-0.01, -0.03, 0.01, 0.02, 0.03]}

df = pd.DataFrame(dados_financeiros)

# Normalizando os dados financeiros
scaler = StandardScaler()
dados_normalizados = scaler.fit_transform(df)

print("Dados normalizados:\n", dados_normalizados)
```



#### 2. Aplicar o PCA

Agora, aplicamos o PCA para reduzir a dimensionalidade dos dados. O PCA nos permitirá encontrar os fatores principais que explicam a maior parte da variação nos retornos das ações.



```python

# Aplicando PCA
pca = PCA(n_components=5)
dados_pca = pca.fit_transform(dados_normalizados)

# Variância explicada por cada componente
print("Variância explicada por cada componente:\n", pca.explained_variance_ratio_)

# Exibindo os componentes principais
print("Componentes principais:\n", dados_pca)
```



Aqui, podemos observar que os componentes principais explicam diferentes porcentagens da variação nos retornos das ações. O objetivo do PCA é identificar essas direções de máxima variação no mercado.

#### 3. Aplicar o Whitening

Após aplicar o PCA, utilizamos o Whitening para garantir que os componentes principais tenham variância unitária e estejam completamente descorrelacionados.



```python

# Aplicando Whitening nos dados projetados pelo PCA
dados_whitened = dados_pca / np.sqrt(pca.explained_variance_)

print("Dados após o Whitening:\n", dados_whitened)
```



Nesse passo, dividimos os componentes principais pela raiz quadrada de seus respectivos autovalores. Isso garante que todas as variáveis resultantes tenham a mesma variância (unitária) e estejam descorrelacionadas.

#### 4. Analisar os resultados

Podemos agora visualizar os resultados do PCA e do Whitening. Vamos traçar um gráfico da variância explicada por cada componente principal (scree plot), bem como comparar os dados antes e depois do Whitening.



```python

# Scree plot da variância explicada
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Explicada Acumulada')
plt.title('Curva de Carga (Scree Plot)')
plt.grid(True)
plt.show()
```



Esse gráfico mostra a variação explicada acumulada pelos componentes principais. Tipicamente, um ponto de "cotovelo" na curva pode indicar quantos componentes principais devemos manter.

#### 5. Considerações sobre a análise

Após a aplicação do PCA e Whitening, podemos observar o seguinte:

1. **Redução de Dimensionalidade**: Identificamos que talvez 2 ou 3 componentes principais explicam a maior parte da variação nos retornos das ações. Isso sugere que, em vez de analisar 5 variáveis (ações), podemos reduzir o problema a 2 ou 3 fatores que influenciam os retornos dessas ações.
   
2. **Descorrelação e Normalização**: O Whitening garante que os componentes principais resultantes não tenham correlação e sejam normalizados, o que é útil para certas aplicações, como em algoritmos de aprendizado de máquina ou para evitar multicolinearidade em modelos estatísticos.

3. **Interpretação Financeira**: No contexto financeiro, os componentes principais podem ser interpretados como fatores de mercado subjacentes. Por exemplo, o primeiro componente principal pode representar uma tendência geral do mercado, enquanto o segundo componente pode representar uma divergência entre ações mais voláteis e menos voláteis.

#### 6. Aplicação em Modelos de Portfólio

No contexto financeiro, PCA pode ser aplicado para construir modelos de portfólio reduzindo a dimensionalidade dos dados de retornos de ativos. Ao identificar os principais componentes que afetam o desempenho das ações, investidores podem concentrar-se nos fatores mais importantes para a construção de seus portfólios, diversificando com base em componentes principais e não em todos os ativos diretamente.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# Gerando um dataset com 4 dimensões e correlação geométrica entre as variáveis
np.random.seed(42)
n_points = 150

# Criando as dimensões com correlação
X1 = np.linspace(-5, 5, n_points)
X2 = X1**2 + np.random.normal(scale=1, size=n_points)
X3 = X1**3 + np.random.normal(scale=1, size=n_points)
X4 = X1**4 + np.random.normal(scale=1, size=n_points)

# Combinando em um dataset
dataset = pd.DataFrame({'X1': X1, 'X2': X2, 'X3': X3, 'X4': X4})

# Visualização do dataset original (quarto eixo como mapa de calor)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(dataset['X1'], dataset['X2'], dataset['X3'], c=dataset['X4'], cmap='viridis')
plt.colorbar(scatter, ax=ax, label='X4')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
plt.title("Dataset Original em 3D com X4 como mapa de calor")
plt.show()

# Aplicando PCA para redução de 4 dimensões para 3
pca_3d = PCA(n_components=3)
dataset_pca_3d = pca_3d.fit_transform(dataset)

# Visualizando a redução para 3 dimensões
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter_pca_3d = ax.scatter(dataset_pca_3d[:, 0], dataset_pca_3d[:, 1], dataset_pca_3d[:, 2], c=dataset['X4'], cmap='viridis')
plt.colorbar(scatter_pca_3d, ax=ax, label='X4')
ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
plt.title("Dataset após PCA (redução para 3 dimensões)")
plt.show()

# Aplicando PCA para redução de 4 dimensões para 2
pca_2d = PCA(n_components=2)
dataset_pca_2d = pca_2d.fit_transform(dataset)

# Visualizando a redução para 2 dimensões
plt.figure(figsize=(8, 6))
scatter_pca_2d = plt.scatter(dataset_pca_2d[:, 0], dataset_pca_2d[:, 1], c=dataset['X4'], cmap='viridis')
plt.colorbar(scatter_pca_2d, label='X4')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title("Dataset após PCA (redução para 2 dimensões)")
plt.show()
```
